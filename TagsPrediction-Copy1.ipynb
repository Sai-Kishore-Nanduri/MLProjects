```python
#Get the frequency distribution of tags in the available
import nltk

keywords = nltk.FreqDist(list_tags)
```

```python
keywords
```

```python
#Get the keywords with frequency >= 100
freq_keywords = keywords.most_common(100)
```

```python
#Get the list of frequent keywords
freq_keywords_list = [i[0] for i in freq_keywords ]
```

```python
freq_keywords_list
```

```python
new_df = df.copy() #Create a copy of dataframe
```

```python
#Define a function to  keep only the most common tags i.e. tags with frequency>=100
def most_common(tags):
    tags_filtered = []
    for i in range(0, len(tags)):
        if tags[i] in freq_keywords_list:
            tags_filtered.append(tags[i])
    return tags_filtered
```

```python
#Filter the most common tags 
new_df['Tags'] = new_df['Tags'].apply(lambda x: most_common(x))
new_df['Tags'] = new_df['Tags'].apply(lambda x: x if len(x)>0 else None) #Replace empty rows with None
```

```python
#Drop Id column
new_df = new_df.drop('Id',axis=1)
```

```python
#Drop the rows with 'None' tags
new_df.dropna(subset=['Tags'],inplace=True)
```

```python
#Check the number of rows dropped
new_df.shape
```

```python
#Remove html tags from the body of the data
from bs4 import BeautifulSoup
new_df['Body'] = new_df['Body'].apply(lambda x: BeautifulSoup(x).get_text())
```

```python
#Define a function to convert text in to lowercase and remove abbrevations
import re
def clean_text(text):
    text = text.lower()
    text = re.sub(r"what's", "what is ", text)
    text = re.sub(r"\'s", " ", text)
    text = re.sub(r"\'ve", " have ", text)
    text = re.sub(r"can't", "can not ", text)
    text = re.sub(r"n't", " not ", text)
    text = re.sub(r"i'm", "i am ", text)
    text = re.sub(r"\'re", " are ", text)
    text = re.sub(r"\'d", " would ", text)
    text = re.sub(r"\'ll", " will ", text)
    text = re.sub(r"\'scuse", " excuse ", text)
    text = re.sub(r"\'\n", " ", text)
    text = re.sub(r"\'\xa0", " ", text)
    text = re.sub('\s+', ' ', text)
    text = text.strip(' ')
    return text
```

```python
#Apply clean_text() to the 'Body' column of the data
new_df['Body'] = new_df['Body'].apply(lambda x: clean_text(x)) 
```

```python
#Use tokenize to convert the text in to list of words
import nltk.tokenize as tkz
import string
tkn = tkz.toktok.ToktokTokenizer()
punct = string.punctuation #Gets all the punctuations as string
```

```python
#Define function to check and remove unnecessary spaces in the text
def getNonEmptyList(x):
    newlist = (item.strip() if hasattr(item, 'strip') else item for item in x)
    return [item for item in newlist if item != '']
```

```python
#Define a function to remove punctuations
def removePunctuation(text):
    list_words   = tkn.tokenize(text)
    regex        = re.compile('[%s]'% re.escape(punct))
    filtered_list= []
    
    for word in list_words:
        if word in freq_keywords_list:
            filtered_list.append(word)
        else:
            filtered_list.append(regex.sub('',word))
            
    nonEmptyList = getNonEmptyList(filtered_list)
    
    return ' '.join(map(str,nonEmptyList))
              
    
```

```python
#Remove punctuations from the text in 'Body'
new_df['Body'] = new_df['Body'].apply(lambda x: removePunctuation(x))
```

```python
#Use snowball stemmer to stem the words of text and stopwords of nltk.corpus to remove stopwords
import nltk
from nltk.stem.snowball import SnowballStemmer
from nltk.corpus import stopwords

stemmer        = SnowballStemmer('english')
stop_words     = set(stopwords.words('english'))

    
```

```python
#Check if 'c' is there in stopwords
check_stop_words = ['c','C']
for i in check_stop_words:
    if i in stop_words:
        stop_words.remove(i)
```

```python
#Define function to step the words in text
def stemWords(mylist):
    
    words         = tkn.tokenize(mylist)
    stemmed_words = []
    
    for word in words:
        
        stemmed   = stemmer.stem(word)
        stemmed_words.append(stemmed)
        
    return ' '.join(map(str, stemmed_words))

#Define function to remove stop words in text
def removeStopWords(mylist):
    
    words         = tkn.tokenize(mylist)
    list_filtered = []
    for word in words:
        
        list_filtered = [word for word in words if not word in stop_words]
    
    return ' '.join(map(str, list_filtered))
        
    
```

```python
# Perform word stemming and remove stop words from the 'Body'
new_df['Body'] = new_df['Body'].apply(lambda x: stemWords(x)) 
new_df['Body'] = new_df['Body'].apply(lambda x: removeStopWords(x)) 
```

```python
#Clean the text in the title column

new_df['Title'] = new_df['Title'].apply(lambda x: str(x))
new_df['Title'] = new_df['Title'].apply(lambda x: clean_text(x)) 
new_df['Title'] = new_df['Title'].apply(lambda x: removePunctuation(x)) 
new_df['Title'] = new_df['Title'].apply(lambda x: stemWords(x)) 
new_df['Title'] = new_df['Title'].apply(lambda x: removeStopWords(x)) 
```

```python
X1  = new_df['Body']
X2  = new_df['Title']
y   = new_df['Tags']
```

```python
#Use multilabel binarizer to do one hot encoding to the tags 
from sklearn.preprocessing import MultiLabelBinarizer 

binarizer   =   MultiLabelBinarizer()
```

```python
#Binarize the tags
y_new       =   binarizer.fit_transform(y)
```

```python
#Use TfidfVectorizer to prepare Tfidf matrix of 'Title' and 'Body'
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer_X1 = TfidfVectorizer(analyzer = 'word',
                                       min_df=0.0,
                                       max_df = 1.0,
                                       strip_accents = None,
                                       encoding = 'utf-8', 
                                       preprocessor=None,
                                       token_pattern=r"(?u)\S\S+",
                                       max_features=1000)

vectorizer_X2 = TfidfVectorizer(analyzer = 'word',
                                       min_df=0.0,
                                       max_df = 1.0,
                                       strip_accents = None,
                                       encoding = 'utf-8', 
                                       preprocessor=None,
                                       token_pattern=r"(?u)\S\S+",
                                       max_features=1000)
```

```python
#Prepare Tfidf matrices of 'Title' and 'Body'
X1_tfidf = vectorizer_X1.fit_transform(X1)
X2_tfidf = vectorizer_X2.fit_transform(X2)
```

```python
X1_tfidf.shape
```

```python
X2_tfidf.shape
```

```python
#Use hstack to stack the tfidf matrices of 'Title' and 'Body' horizontally
from scipy.sparse import hstack

X_tfidf = hstack([X1_tfidf,X2_tfidf])
```

```python
X_tfidf.shape
```

```python
#Split the data into train and test with test data size 0.2(Train-80% and Test-20%)
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y_new, test_size = 0.2, random_state = 45)
```

```python
#Create a function to print the F1 score based on the model
from sklearn.metrics import f1_score
def print_score(y_pred, clf):
    print("Clf: ", clf.__class__.__name__)
    print("F1 score: {}".format(f1_score(y_test, y_pred,average = 'micro')))
    print("---")    
```

```python
#Fit a model using Logistic Regression with OneVsRestClassifier 
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier
lr   = LogisticRegression()
clf  = OneVsRestClassifier(lr)
```

```python
#Fit the model and predict 'Tags' for Test data
clf.fit(X_train,y_train)
y_pred_lr  = clf.predict(X_test)
```

```python
#Calculate f1-score
print_score(y_pred_lr, lr)
```
